```python
import pandas as pd
```

```python
def combine_df(num):
    df1 = pd.DataFrame()
    dt1 = pd.read_excel('./performance/dt_perf_data'+num+'.xlsx')
    nn1 = pd.read_excel('./performance/nn_perf_data'+num+'.xlsx')
    knn1 = pd.read_excel('./performance/knn_perf_data'+num+'.xlsx')
    svm1 = pd.read_excel('./performance/svm_perf_data'+num+'.xlsx')
    xgb1 = pd.read_excel('./performance/xgb_perf_data'+num+'.xlsx')
    df1=pd.concat([dt1,nn1,knn1,svm1,xgb1], axis=0)
    df1 = df1[['acc','cv_score','train_roc','test_roc','recall_score','f1_score','run_time','name']]
    return df1

```

# Data 1

```python
# load result df
df = combine_df('1')
df[df.name.str.contains("test")]
```
